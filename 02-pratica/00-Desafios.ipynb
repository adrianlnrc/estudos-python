```python
# 01 - Classificador de temperatura
while True:
    try:
        tempAtual = int(input("Hoje esta quantos graus?:"))
        if  tempAtual > 28:
            print(f"o clima esta quente, {tempAtual} graus")
        elif tempAtual >= 15:
            print(f"o clima esta Agradavel, {tempAtual} graus")
        else:
            print(f"Hoje esta um dia frio, com {tempAtual} graus")
        break
    except ValueError:
        print("Digite somente os graus (numeros)")
```

```python
# 02 - Contagem de lançamento da nasa While e For 
import time
contagem = 10

# While
while contagem  >= 0:
    if contagem == 10:
        print("Inicio do Lançamento")
    elif contagem == 0:
        print('Lançamento')
    else:
        print(contagem)
    contagem -= 1
    time.sleep(1)

# For
for i in range(10,-1,-1):
    if i == 10:
        print("Inicio Lançamento")
    elif i == 0: 
        print("Lançamento")
    else:
        print(i)

list(range(10, -1, -1))
```

```python
# 03 - Carrinho de compras
listaCompras = []
for novoItem in range(1,6):
    novoItem = input("Digite um novo item pra lista de compras")
    listaCompras.append(novoItem)

n = 1
for item in listaCompras:
    print(f"item {n}: {item}")
    n += 1
```

```python
# 05 - Estoque de Produtos 
produtos = {
    "maça": 7.90,
    "pera": 20.90,
    'banana': 8.90,
}

for produto,preco in produtos.items():
    print(f"Produto: {produto} | Preço: R$ {preco:.2f}")

soma = 0
for preco in produtos.values():
    soma += preco
print(f'{soma:.2f}')

for produto,preco in produtos.items(): 
    if preco > 20.0:
        print(f'o preco do produto {produto}: é {preco}')
```

```python
# 06 - Tabuada iterativa

n = int(input("Digite um numero que deseja ver a tabuada"))

for x in range(1,11):
    print(f'{n} x {x} = {n * x}')
```
