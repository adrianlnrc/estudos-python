## Mascaras de substituição .format

**Três maneiras de ser feito**
- %s para str e %d para int
- .format ao final do print()
- f antes das aspas dentro do print() e usando {}

```python
# 1. %s 
nome = "Adrian"
idade = 20
print("nome: %s e idade %d"%(nome,idade))

# 2. .format
print("nome {0} e idade {1}".format(nome,idade))

# 3. Mais utilizado
print(f'nome {nome} e idade {idade}')
```

### Estruturas condicionais 

```python
"""" Sao estruturas que mudam o rumo do programa de acordo
com as condiçoes, exemplo if idade >= 18: "voce é maior de idade"
else: "ainda é menor de idade"
"""
idade = int(input("digite a sua idade"))
# if é a primeira condição, se verdadeira ira executar o bloco de instrução  
if idade > 18:
    print("voce ja é maior de idade")
# elif é a segunda condição, caso o if nao seja verdadeira
elif idade == 18: 
    print("Voce acabou de fazer 18 anos")
# else é a ultima condição, quando nenhuma anterior é acionada
else: 
    print("voce ainda é menor de idade") 

nome = "Bob"
# Tambem é possivel utiliza dentro de condicionais (AND/OR)

if idade >= 18 and nome == "Bob":
    print(f"Seu nome é {nome}, e voce tem {idade} anos")
```

### Listas/Array

Lista é um objeto que permite guardar diversos elementos de maneira indexada

```python
lista = [1,2,3,4,5,6,7]

# Mostra todos os Métodos/comandos diante de uma Lista
print(dir(list))

# Adiciona elemento no final da lista
lista.append(8)
print(lista)

#removendo o elemento pelo indice (pop) ou por incidencia (remove)
lista.pop(1)
print(lista)
lista.remove(4)
print(lista)

# Mostra o index de um elemento
print(lista.index(3))
```

### Dicionarios (ou objeto)

- Utilizam chave valor para armazenar informações 

- Dicionarios possibilitam acessar em uma variável mais de um valor (atributo), permitindo agrupar informações em uma única variável

```python
# Exemplo de Criação de Dicionário e consultas de dados 

# Informações do jogador, criando um dicionário
jogador = {"nome": "Bob",
           "nível": 1,
           "vida": 100,
           "exp": 0,
           "dano": 5,  
        }
# Informações de inimigos, criando uma lista de dicionários
inimigos = [
    {"nome": "Aldeão", "dano": 0, "vida": 100, "exp": 2},
    {"nome": "Dragão", "dano": 25, "vida": 150, "exp": 20},
    {"nome": "Monstro", "dano": 30, "vida": 200, "exp": 30},
    {"nome": "Larva", "dano": 15, "vida": 75, "exp": 15},
]
# Tipo do objeto type(dict), quantidade de chaves valores len("nome_dicionario") e comandos disponiveis em dicionários
print(type(jogador))
print(len(jogador)) 
print(dir(jogador))

# Consultando dados via .get(nome) e chaves via .keys() 
print(jogador.get('nome'))
print(jogador.keys())

# Loop mostrando cada elemento do dicionario inimigos
for inimigo in inimigos:
    print(inimigo['nome'])
    print(inimigo['dano'])
```

```python
# Manipulação de dados utilizando Dicionários

carros = {'nome': 'bmw m3',
          'potência': 500,
          'tração': 'traseira',
          'anos fabricados': ['2017','2018','2020'],
          }
# Alterando o valor de potência
carros['potência'] = 510
print(carros)

# Adicionando novos dados a um dicionário, Para que seja possível adicionar mais de um valor a uma chave, é necessario seguir uma sintaxe de lista
carros['anos fabricados'].append('2022')
print(carros.values())
```

## Conjuntos (set)
- parece uma lista mas nao aceita valores repetidos e nao é indexável

```python
c1 = {1,2,3,4,5}
c2 = {1,2,3,4,5,6,7}
# União: juntando todos os dados em comum entre os conjuntos
print(c1.union(c2))

# Interseção: mantem so os elementos em comum em ambos conjuntos
print(c1.intersection(c2))

# Utilizando .update() é possivel alterar o elementos de forma permanente
c1.update(c2)

# Verificacao se c2 esta contido ou nao dentro de c1, resultado True por conta de 
c2 <= c1
```

## Funções 
**é um bloco de código que pode ser executados e reutilizados mais de uma vez**
- Um exemplo seria um sistema que soma varias vezes dois numeros, é possivel criar uma função que calcule a soma somente mudando os seus parâmetros 
- Nem todas as funções retornam um valor 

**Python ja possui por padrão varias funções pre-definidas que executam um determinado comando**
 - Como *print()* que mostra na tela tudo que é passado como parâmetro dentro dos ()
 - Como len() para mostrar o tamanho do objeto
 - Outro exemplo seria adicionar a biblioteca random e utilizar os comandos: *random.randint(0,5)* para gerar um numero pseudo-aleatório no intervalo passado como parâmetro

```python
# Criando Funções personalizadas 

def BemVindo():
    print('Seja bem vindo')

print(BemVindo())

def calculadora(num1,num2,operacao):
    resultado = 0

    if operacao == 'soma':
        resultado = num1 + num2
    elif operacao == 'subtracao':
        resultado = num1 - num2
    elif operacao == 'multiplicacao':
        resultado = num1 * num2
    else:
        resultado = num1 // num2

    return resultado

num1 = int(input('digite o 1 numero'))
num2 = int(input('digite o 2 numero'))
op = input("qual operacao? (soma),(subtracao),(multiplicacao),(divisao)")
op.lower()

print(calculadora(num1,num2,op))
```

## Importar bibliotecas 

- Bibliotecas sao um conjunto de parâmetros e funções prontas juntas em um pacote 

- Funciona da seguinte forma, de acordo com a evolução de um projeto é necessario novas funcionalidades além das bibliotecas padrões do python, entao é importado novas funcionalidades de funções nativas ou externas do python

**Em python podemos Importar a biblioteca inteira utilizando somente o import e o nome de referência, ou podemos importar somente uma parte de dentro daquela biblioteca, utilizando from (nome da biblioteca) import (nome da funcao)**

```python
from math import pi
import random

print(pi) # numero de pi importado da biblioteca math

lista = []
for x in range(0,20):
    item = random.randint(0,5)
    lista.append(item)
print(lista)
```
