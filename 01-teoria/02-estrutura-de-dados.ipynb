### Listas

- São o equivalente a arrays, uma lista é um objeto que permite guardar diversos dados dentro dele de forma organizada e indexada em um espaço só da memória, permitindo leitura de dados rapidamente, ótimo para **acesso aleatório**

- Inserçao e deleção em listas possui tempo de complexidade **O(1)** quando inserido no final da lista, mas no pior caso possui **O(n)**, pois ex: em uma inserção no meio terá de mover a lista inteira a direita para "abrir espaço".


```python
lista_de_numeros = [1,2,3,4,5,6,7,8,9]

print(lista_de_numeros)
```

### Lista encadeada 

- A lista encadeada funcina da seguinte forma, ela nao possui um índice, cada elemento possui o endereço do próximo elemento, **nao** sendo sequencial na memória

- Leitura da Lista encadeada por nao possuir índice é necessario que passe por todos os elementos ate encontrar o elemento do meio por exemplo, sendo **O(n)** para acesso.

- Inserção e deleção no início em lista encadeadas possui tempo de complexidade O(1) e no final, assumindo que você guarda um ponteiro para o último nó, enquanto no meio a lista precisa navegar até posição **O(n)**.

### Pilhas 

- nada mais sao do que listas em que a inserção e a remoção dos elementos acontecem na mesma extremidade, seguindo o padrão **first in, last out**

```python
pilha = [10,20,30]

# Insere um elemento no final da lista
pilha.append(40)
print(pilha)

# Remove o último valor da lista, quando nao passado nenhum parâmetro
pilha.pop()
print(pilha)

# Tamanho da pilha através de len(parâmetro) e o topo
print(len(pilha))
```
