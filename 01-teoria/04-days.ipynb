```python
# Day 5 Lists 

lista = ["a","d","r","i","a","n"] # forma convencional de criar uma lista 
print(f"lista: {lista}")
lista_vazia = list()
print(len(lista_vazia))

lista_compras = ['tomate','acucar','pao']
print(f"numero de itens na lista: {len(lista_compras)}")

# Listas podem ter diferentes tipos de dados 
lst = ['adrian',33,True, {'idade': 22, 'faculdade': 'engenharia de software'}]
print(type(lst))

# Acessar valor pelo índice da lista, que sempre inicia em 0
primeiro_item = lista_compras[0]
segundo_item = lista_compras[1]
print(f"primeiro item: {primeiro_item}, segundo item: {segundo_item}")
# Ultima posição da lista
ultima_posicao = len(lista_compras) - 1 # porque len pega o a quantidade de itens, e -1
print(ultima_posicao)

# iterable unpacking (*)
lista = ['item1','item2','item3','item4','item5']
# Jeito antigo utilizando Slicing, utilizando índices
rest = lista[3:]
print(rest)

# Jeito moderno utilizando o asterisco, pegando o resto da lista e atribuindo no rest, * sendo um Imã
first,second,third, *rest = lista # pegando o final da lista 

first,*mid, last = lista # pegando o meio da lista 
print(mid)

# "Slicing" items
# Sendo [0:1:2] 0 = inicio 1 = final(exclusivo) 2 = tamanho do passo

frutas = ['banana','mamao','uva','abacaxi','maça']

todasFrutas = frutas[0:] # inicia na posicao 0 sem definir o final, pegando todos os items 
print(todasFrutas)

uva_abacaxi = frutas[2:4] # inicia na posicao 2 (entra no resultado) e termina na posicao 4 (nao entra)
print(uva_abacaxi)

frutasPares = frutas[::2] # Pulando de 2 em 2 
frutasInvertidas = frutas[::-1] # iniciando do final da lista de 1 por 1
print(frutasPares, frutasInvertidas)

# Exercicio
dados_site = ['Serviço 1', 'vazio', 'Serviço 2', 'vazio', 'Serviço 3', 'vazio']
servicos = dados_site[::2]
print(servicos)

# Modificando listas 
frutas = ['banana','mamao','uva','abacaxi','maça']

frutas[0] = 'banana da terra'
print(frutas)
ultimoIndice = len(frutas) -1
frutas[ultimoIndice] = 'maça gala'
print(frutas)
```
